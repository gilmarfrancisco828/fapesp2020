```python
import numpy as np
from methods import *
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def f(y, x):
    return -y + x
def analitica(x):
    return (np.exp(x)*x - np.exp(x)+2)/np.exp(x)
```

$y=\frac{e^xx-e^x+2}{e^x}$

```python
a, b = 0., 1. # intervalo
y_a = 1. #condicao inicial
hs = [10**(-i) for i in range(3, 9)]
print(hs)
```

```python
# ConvergÃªncia Para Euler
MSEs = []
AE = []
for h in hs:
    # criando a malha
    x = np.linspace(a, b, int((b - a+h)/h))
    
    # solucao
    res_euler = euler(f, y_a, x)
    res_analitica = analitica(x)
    erro_euler = res_analitica - res_euler
    MSE_euler = np.sum(np.power(erro_euler, 2)) / len(x)
    
    AE.append(np.sum(np.absolute(erro_euler)))
    MSEs.append(MSE_euler)
    
```

```python
plt.figure(dpi=100, figsize=(16, 6))
plt.plot(hs, MSEs, 'b', label='Erro', marker='o')
plt.title('ConvergÃªncia para Euler MSE')
plt.legend(loc='best')
plt.grid()
plt.yscale("log")
plt.xscale("log")
plt.xlabel('h')
plt.ylabel('Erro') 
plt.xticks(hs)
plt.show()

plt.figure(dpi=100, figsize=(16, 6))
plt.plot(hs, AE, 'b', label='Erro', marker='o')
plt.title('ConvergÃªncia para Euler Erro Absoluto')
plt.legend(loc='best')
plt.grid()
plt.yscale("log")
plt.xscale("log")
plt.xlabel('h')
plt.ylabel('Erro') 
plt.xticks(hs)
plt.show()

plt.figure(dpi=100, figsize=(16, 6))
plt.plot(hs[2:], AE[2:], 'b', label='Erro', marker='o')
plt.title('ConvergÃªncia para Euler Erro Absoluto')
plt.legend(loc='best')
plt.grid()
plt.yscale("log")
plt.xscale("log")
plt.xlabel('h')
plt.ylabel('Erro') 
plt.xticks(hs[2:])
plt.show()
```
