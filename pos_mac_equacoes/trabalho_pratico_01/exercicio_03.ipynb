```python
n = 101
a, b = 0., 1. # intervalo
y_a, y_b = 1., 0. #condicao inicial
hs = [10**(-i) for i in range(3, 9)]
t1 = np.linspace(a, b, n)
print(t1[1] - t1[0])
```

```python
def u(x):
    return 2*x
def v(x):
    return -1
def w(x):
    return -x
```

```python
sol = fd_central( u(t1), v(t1), w(t1), t1, y_a, y_b, method="central")
```

```python
plt.figure(figsize=(20,10))
plt.plot(t1, sol, label='Diferen√ßas Finitas Centrada', marker='o')
plt.legend(loc='best')
plt.grid()
plt.show()
```
