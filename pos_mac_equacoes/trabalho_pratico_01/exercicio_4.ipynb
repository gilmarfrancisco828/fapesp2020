```python
import numpy as np
from fd_central import *
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
n = 6
# n = [6, 11, 26]
a, b = 0, 1
y_a, y_b = 0, np.e
t1 = np.linspace(a, b, n)
print(t1, t1[1] - t1[0])
```

```python
def u(x):
    return np.exp(x)*(x**2 + 1)
def v(x):
    return -x
def w(x):
    return 1
```

```python
xfd1 = fd_central( u(t1), v(t1), w(t1), t1, y_a, y_b, method="foward")
xfd2 = fd_central( u(t1), v(t1), w(t1), t1, y_a, y_b, method="central")
xfd3 = fd_central( u(t1), v(t1), w(t1), t1, y_a, y_b, method="backward")
```

```python
plt.figure(figsize=(20,10))

plt.plot(t1, xfd1, label='Dif. centrada e avanada', marker='o')
plt.plot(t1, xfd2, label='Dif. centrada e centrada', marker='o')
plt.plot(t1, xfd3, label='Dif. centrada e atrasada', marker='o')
plt.grid()
plt.legend(loc='best')
plt.show()
```
