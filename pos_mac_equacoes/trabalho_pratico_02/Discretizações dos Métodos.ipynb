```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg' # display plots in SVG format
path = '/home/parafuso828/Documents/UNESP/4_SEM_1/Métodos Numéricos para Equações/trabalho_pratico_02/img/'
plt.figure(figsize=(8,4))
plt.plot([0,2],[0,0],'k')
plt.plot([1,1],[0,1],'k')
plt.plot([0,1,2,1],[0,0,0,1],'ko',markersize=10)
plt.text(1.1,0.1,'T[i,j]')
plt.text(0.1,0.1,'T[i-1,j]')
plt.text(1.1,1.1,'T[i,j+1]')
plt.text(2.1,0.1,'T[i+1,j]')
plt.xlabel('x')
plt.ylabel('t')
plt.axis('equal')
plt.yticks([0.0,1.0],[])
plt.xticks([0.0,1.0],[])
plt.title('Método Explícito',fontsize=12)
plt.axis([-0.5,2.5,-0.5,1.5]);
plt.savefig(path+'disc_e_explicito')
```

```python
plt.figure(figsize=(8,4))
plt.plot([0,2],[1,1],'k')
plt.plot([1,1],[0,1],'k')
plt.plot([0,1,2,1],[1,1,1,0],'ko',markersize=10)
plt.text(1.1,0.1,'T[i,j]')
plt.text(0.1,1.1,'T[i-1,j+1]')
plt.text(1.1,1.1,'T[i,j+1]')
plt.text(2.1,1.1,'T[i+1,j+1]')
plt.xlabel('x')
plt.ylabel('t')
plt.axis('equal')
plt.yticks([0.0,1.0],[])
plt.xticks([0.0,1.0],[])
plt.title('Método Implícito',fontsize=12)
plt.axis([-0.5,2.5,-0.5,1.5]);
plt.savefig(path+'disc_e_implicito')
```

```python
plt.figure(figsize=(8,4))
plt.plot([0,2],[0,0],'k')
plt.plot([0,2],[1,1],'k')
plt.plot([1,1],[0,1],'k')
plt.plot([0,1,2,0,1,2],[0,0,0,1,1,1],'ko',markersize=10)
plt.text(0.1,0.1,'T[i-1,j]')
plt.text(1.1,0.1,'T[i,j]')
plt.text(2.1,0.1,'T[i+1,j]')
plt.text(0.1,1.1,'T[i-1,j+1]')
plt.text(1.1,1.1,'T[i,j+1]')
plt.text(2.1,1.1,'T[i+1,j+1]')
plt.xlabel('x')
plt.ylabel('t')
plt.axis('equal')
plt.yticks([0.0,1.0],[])
plt.xticks([0.0,1.0],[])
plt.title('Crank-Nicolson',fontsize=12)
plt.axis([-0.5,2.5,-0.5,1.5]);
plt.savefig(path+'disc_crank_nicolson')
```
